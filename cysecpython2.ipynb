## **O que é engenharia social?**
Bem como na última edição deste projeto, novamente iremos abordar como tema central a engenharia social como agente causador dos principais ataques cibernéticos, mas antes deve-se re-explicar o que é a engenharia social, ou como alguns chamam, "a arte de enganar".

Em resumo, esta é a técnica mais utilizada por hackers, crackers e criminosos em geral, na qual consiste em um atacante manipular, persuadir ou enganar uma vítima de tal forma que a mesma forneça informações pessoais, credenciais de acesso ou até mesmo acesso lícito ou ilícito de seus dispositivos informáticos sem ter consciência de tal ato.

# **TikTok**

Lançado em 2014 na China com o nome de Douyin e em 2017 vendido à empresa chinesa ByteDance por cerca de 800 milhões de dólares e lançado no mercado internacional no mesmo ano, porém com o nome de Musical.ly, no ano de 2018 a empresa encerrou o aplicativo e lançou o TikTok, com mecânicas e propostas novas e/ou melhora de funções existentes no antigo aplicativo, e moveu todos os usuários da antiga plataforma para a nova, fazendo com que o "novo" aplicativo recebesse mais de 500 milhões de usuários até a metade do mesmo ano.

Através do gráfico abaixo, pode-se perceber uma gradativa evolução no número de downloads do aplicativo por semestre e que, se somarmos os valores apresentados a seguir juntamente com a quantidade de downloads feitos no primeiro semestre deste de 2020, que é de mais de 500 milhões (maior que qualquer semestre anterior), obtemos um total de mais de 2 bilhões de downloads.

```
#Importando bibliotecas utilizadas
import pandas as pd
import matplotlib.pyplot as plt

#Importando dados
downloads = pd.read_csv('https://raw.githubusercontent.com/sidneypepo/cysecpython2/main/files/downloads.csv', usecols=['downloads'])
semestres = ['2017.1', '2017.2', '2018.1', '2018.2', '2019.1', '2019.2']

#Plotando gráfico
fig, ax1 = plt.subplots()
ax1.plot(semestres, downloads, color="#ff2929")
plt.style.use('dark_background')
plt.style.context('dark_background')
fig.set_size_inches(9, 5)
 
#Configurações do gráfico
plt.title('Downloads do TikTok de 2017 até 2019\n', fontsize=20)
plt.xlabel('Semestres', fontsize=15)
plt.ylabel('Total de downloads\'s (em centenas de milhões)', fontsize=15)
plt.rc('xtick', labelsize=15)
plt.rc('ytick', labelsize=15)
plt.grid(True)
plt.show()
 
#Fonte
print("\nFONTE: \n<CRAIG CHAPPLE. TikTok Crosses 2 Billion Downloads After Best Quarter For Any App Ever. 2020. \nDisponível em: https://sensortower.com/blog/tiktok-downloads-2-billion. \nAcesso em: 22 out. 2020.>")
```

No entanto, foi-se percebido certa dificuldade para "entrar" no aplicativo, pois o mesmo realizava modificações em seu sistema para impedir qualquer tentativa de descompilar o aplicativo, como afirma o pesquisador 'bangorlol' (usuário da rede social Reddit): 

  > "Eles possuem inúmeras proteções diferentes para evitar que você consiga fazer engenharia reversa ou debugue o aplicativo. O comportamento do aplicativo muda rapidamente se perceberem que você está tentando descobrir o que eles estão fazendo. Também há alguns trechos no código da versão do Android que os permite baixar arquivos zip remotamente, descompactando-os e executando o referido binário. Não há nenhuma razão para um aplicativo mobile necessitar legitimamente desta funcionalidade. [...] eles não querem que você saiba quanta informação eles estão coletando sobre você, e as falhas de segurança de todos esses dados em massa um único lugar são enormes. Eles criptografam todas as solicitações de análise com um algoritmo que muda a cada atualização (no mínimo, as chaves mudam) apenas para que você não possa ver o que estão fazendo".

Ainda na pesquisa do 'bangorlol', também foi descoberto que o TikTok coleta: 
  * Hardware do celular (tipo de processador, número telefônico, ID de hardwares, dimensão de tela, DPI, uso de memória, uso de disco, etc.);
  * Outros aplicativos que você instalou;
  * Tudo relacionado à rede (IP, IP local, endereço MAC do roteador, seu endereço MAC, nome da rede Wi-Fi);
  * Se seu dispositivo está com acesso root/desbloqueado;
  * Algumas variantes do aplicativo possuíam ping de GPS habilitado, executado a cada aproximadamente 30 segundos;
  * Eles configuram um servidor proxy local em seu aparelho para "transcodificação de mídia", mas isso pode ser abusado facilmente por não possuir autenticação;

Além das pesquisas diretamente realizadas no códido da mídia social, o analista de redes Luis Agner Claro, digital influencer e dono do canal [CPU não é Gabinete](https://www.youtube.com/channel/UCI9J4UNaNzSoeaVxYMyAn2A), fez uma análise a respeito da maneira em que as informações do TikTok trafegadas na rede são feitas e para onde as mesmas vão. Com isso, apenas foi instalado o TikTok no aparelho celular e então o aplicativo foi totalmente fechado e então o telefone foi conectado na rede criada pelo roteador e durante o tempo que este aparelho estava conectado, foi-se gerado um arquivo de log, ou seja, um arquivo contendo todas as informações que foram trafegadas entre o dispositivo e o ponto de acesso à internet. # **"Mas qual a diferença da coleta de dados do TikTok para as outras redes sociais?"**

Com base nas análises anteriores, podemos notar nitidamente comportamentos no mínimo questionáveis a respeito do aplicativo chinês, "mas qual a diferença da coleta de dados do TikTok para as outras redes sociais"? Retornando às informações divulgadas no Reddit pelo usuário 'bangorlol', há um trecho em que o mesmo faz a seguinte afirmação:

  > "Para isso que valesse a pena, fiz engenharia reversa nos aplicativos Instagram, Facebook, Reddit e Twitter. Eles não coletam nada semelhante à quantidade de dados que o TikTok, e eles não tentam esconder o que fazem exatamente com essas informações que são enviadas, diferentemente do TikTok. É como comparar um copo de água com o oceano - não há como compará-los".

Diante da informação apresentada anteriormente, nós do projeto CySecPython realizamos pesquisas e testes que comprovam a tese do 'bangorlol', porém antes de apresentarmos as informações obtidas, devemos primeiramente explicar o que são *trackers* (rastreadores):

  * Basicamente, os *trackers* (ou rastreadores) são códigos ocultos que ficam em funcionamento durante sua estadia em algum serviço ou website. Suas funções vão desde coletar seu comportamento e modo de interação com os recursos da página até salvarem histórico e cookies de seu navegador e venderem as informações obtidas sobre você para empresas (normalmente de publicidade).

Com o conceito explicado, vamos ao teste: primeiramente instalamos em um navegador web a extensão/add-on uBlock Origin, que tem como função bloquear e alertar a respeito dos trackers existentes na página, então entramos na página principal de cada uma das redes sociais mencionadas anteriormente (Instagram, Facebook, Reddit, Twitter e TikTok) e permanecemos em cada uma por cerca de 3 minutos cada.
